```python
# raters
raters = ['Jon', 'Josh', 'Kevin', 'Nate']
```

```python
import pandas as pd

raters = ['Jon', 'Josh', 'Kevin', 'Nate']

c = pd.read_csv('data/MKM_Kevin_commons.csv')
u = pd.read_csv('data/MKM_Kevin_uncommons.csv')
eos = pd.concat([c, u], ignore_index=True)
eos = eos[['Name', 'Color', 'Rarity']]

for rater in raters:
    c = pd.read_csv('data/MKM_' + rater + '_commons.csv')
    u = pd.read_csv('data/MKM_' + rater + '_uncommons.csv')
    temp = pd.concat([c, u], ignore_index=True)
    eos[rater] = temp[rater]

tier_num = {'F': 0., 'G':0., 'G+':0., 'D-': 1., 'D': 2., 'D+': 3., 
            'C-': 4., 'C': 5., 'C+': 6., 'B-': 7., 'B': 8., 'B+': 9., 
            'A-': 10., 'A': 11., 'A+': 12.}

for rater in raters:
    eos[str(rater) + '_num'] = eos[rater].map(tier_num)

# adding in the MKM data the hard way to make sure the names match 
l = pd.read_csv('data/MKM_tier_list.csv')
names_eos = eos['Name'].tolist()
tiers = ['Unknown'] * len(names_eos)

for i, name in enumerate(names_eos):
    if name in l['Name'].tolist():
        index = l['Name'].tolist().index(name)
        tiers[i] = l.loc[index, 'Tier']

eos['Tier'] = tiers

eos['Tier_num'] = eos['Tier'].map(tier_num)
for rater in raters: 
    eos[str(rater) + '_error'] = abs(eos['Tier_num'] - eos[str(rater) + '_num'])
    
eos
```

## And The Winner Is... Jon

```python
for rater in raters:
    error = eos[f'{rater}_error'].sum()
    print(f'Error for {rater}: {error}')
```

## Biggest Misses


```python
for rater in raters:
    temp = eos.copy()
    temp = temp.sort_values(by=f'{rater}_error', ascending=False)
    print(temp[['Name', 'Tier', rater]][0:5])
```

```python
for rater in raters:
    temp = eos.copy()
    other_raters = [x for x in raters if x != rater]
    temp[f'{rater}_hot'] = abs(((temp[[f'{x}_num' for x in other_raters]].sum(axis=1)) / (len(other_raters))) - temp[f'{rater}_num'])
    temp = temp.sort_values(by=f'{rater}_hot', ascending=False)
    top_5_cols = ['Name', 'Tier'] + [f'{x}' for x in raters]
    top_5 = temp[top_5_cols].head(5)
    print(f'Top 5 for {rater}:')
    print(top_5)
```
